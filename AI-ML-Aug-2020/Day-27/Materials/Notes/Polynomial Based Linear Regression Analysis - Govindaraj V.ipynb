```python
x=data.iloc[:,:1]
y=data.iloc[:,-1:]
print(x.head(1))
print()
print(y.head(1))
```

```python
from sklearn.linear_model import LinearRegression
lin_reg=LinearRegression()
lin_reg.fit(x,y)
```

```python
plt.scatter(x,y)
plt.plot(x,lin_reg.predict(x),color='green')
plt.title('Base Linear regression model Regression line graph')
plt.xlabel('Age')
plt.ylabel('Claim per policy')
```

# Polynomial Based LR

```python
from sklearn.preprocessing import PolynomialFeatures
poly=PolynomialFeatures(degree=5)
x_poly=poly.fit_transform(x)
poly_reg=LinearRegression()
poly_reg.fit(x_poly,y)
plt.scatter(x,y,color='red')
plt.plot(x,lin_reg.predict(x),color='green')
plt.plot(x,poly_reg.predict(x_poly),color='blue')
plt.title('Polynomial based-Linear regression model Regression line graph')
plt.xlabel('Age')
plt.ylabel('Claim per policy')
```

```python
lr_pred=lin_reg.predict(np.array([[15]]))
poly_predi=poly_reg.predict(poly.fit_transform(np.array([[15]])))
print('Linear Regression oucome\n',lr_pred)
print('Poly based LR prediction\n',poly_predi)
```
