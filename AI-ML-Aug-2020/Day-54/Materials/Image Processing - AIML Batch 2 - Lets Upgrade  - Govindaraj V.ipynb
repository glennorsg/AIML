```python
# BGR to GRAY

bgr_gray=cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
cv2.imshow('BGR-GRAY',bgr_gray)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

# cv2.warpAffine

```python
import cv2
import numpy as np
img=cv2.imread('C:/Users/guruc/Documents/AIML/dataset/nature.jpg',0)
```

```python
rows,cols=img.shape
X=np.float32([[1,0,100],[0,1,50]])
trans=cv2.warpAffine(img,X,(cols,rows))
cv2.imshow('WarpAffine',trans)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
img.shape
```

```python
X=cv2.getRotationMatrix2D((cols/2,rows/2),90,1)
rorate=cv2.warpAffine(img,X,(cols,rows))
re_size=cv2.resize(rorate,None,fx=3,fy=3,interpolation=cv2.INTER_CUBIC)
cv2.imshow('Rotated_Image',re_size)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

# cv2.warpPerspective

```python
import cv2
import numpy as np
img=cv2.imread('C:/Users/guruc/Documents/AIML/dataset/nature.jpg')
```

```python
rows,cols,ch=img.shape
```

```python
pos1=np.float32([[56,65],[365,87],[16,387],[289,379]])
pos2=np.float32([[0,0],[300,0],[0,300],[300,300]])

X=cv2.getPerspectiveTransform(pos1,pos2)

rotate_rgb_img=cv2.warpPerspective(img,X,(300,300))

plt.subplot(121),plt.imshow(img),plt.title('Actual Input Image')

plt.subplot(122),plt.imshow(rotate_rgb_img),plt.title('Output')

plt.show()
```

# Smoothing

```python
import cv2
import numpy as np
img=cv2.imread('C:/Users/guruc/Documents/AIML/dataset/nature.jpg')
```

```python
kernal=np.ones((5,5),np.float32)/45
```

```python
kernal
```

```python
blur_image=cv2.filter2D(img,-1,kernal)

plt.subplot(121),plt.imshow(img),plt.title('Actual Input Image')

plt.subplot(122),plt.imshow(blur_image),plt.title('Averaging blur image')

plt.show()
```

# cv2.GaussianBlur

```python
gauss_blur=cv2.GaussianBlur(img,(5,5),0)
plt.subplot(121),plt.imshow(img),plt.title('Actual Input Image')

plt.subplot(122),plt.imshow(gauss_blur),plt.title('Gauss blur image Output')

plt.show()
```
